# 1) Image Classification Dataset

```python
import tensorflow as tf
import matplotlib.pyplot as plt

(x_train, y_train), _ = tf.keras.datasets.cifar10.load_data()
class_names = ['airplane','automobile','bird','cat','deer','dog','frog','horse','ship','truck']

fig, axes = plt.subplots(1, 3, figsize=(9, 3))
for i, ax in enumerate(axes):
    ax.imshow(x_train[i])
    ax.set_title(class_names[y_train[i][0]])
    ax.axis('off')
plt.suptitle("CIFAR-10 — Image Classification")
plt.show()
```

# 3) Instance Segmentation Dataset

```python
!pip install -q datasets

from datasets import load_dataset
import matplotlib.pyplot as plt

ds = load_dataset("timm/oxford-iiit-pet", split="train", trust_remote_code=True)

fig, axes = plt.subplots(1, 3, figsize=(10, 4))
for i, ax in enumerate(axes[:3]):
    sample = ds[i]
    ax.imshow(sample['image'])
    ax.set_title(sample.get('label', ''))
    ax.axis('off')
plt.suptitle("Oxford Pets — Segmentation")
plt.show()
```

# 4) Face Recognition

```python
from sklearn.datasets import fetch_lfw_people
import matplotlib.pyplot as plt

lfw = fetch_lfw_people(min_faces_per_person=70, resize=0.5)

fig, axes = plt.subplots(1, 3, figsize=(9, 4))
for i, ax in enumerate(axes):
    ax.imshow(lfw.images[i], cmap='gray')
    ax.set_title(lfw.target_names[lfw.target[i]], fontsize=9)
    ax.axis('off')
plt.suptitle("LFW — Face Recognition")
plt.show()
```

# 7) Optical Character Recognition

```python
from datasets import load_dataset
import matplotlib.pyplot as plt

ds = load_dataset("priyank-m/IAM_words_data", split="train", streaming=True)

fig, axes = plt.subplots(1, 3, figsize=(10, 3))
for ax, sample in zip(axes, ds.take(3)):
    ax.imshow(sample['image'], cmap='gray')
    ax.set_title(f"\"{sample['label']}\"", fontsize=12)
    ax.axis('off')
plt.suptitle("IAM Handwriting — OCR")
plt.show()
```
